```python
import requests
import json
from openai import AzureOpenAI

openai_endpoint = ""
openai_key = ""

search_service_name = ""
index_name = ""
admin_api_key = ""

client = AzureOpenAI(
  api_key = openai_key,  
  api_version = "2024-02-01",
  azure_endpoint = openai_endpoint
)
```

# Simple Hybrid query

```python
# Getting embedding of the question

question = "What are contraindications?"

response = client.embeddings.create(
    input = question,
    model= "text-embedding-ada-002"
)

# Define the URL and headers
url = f'https://{search_service_name}.search.windows.net/indexes/{index_name}/docs/search?api-version=2023-11-01'
headers = {
    'Content-Type': 'application/json',
    'api-key': admin_api_key
}

# Define the request body
payload = {
    "vectorQueries": [
        {
            "vector": response.data[0].embedding,
            "fields": "chunkVector",
            "kind": "vector",
            "k": 50
        }
    ],
    "search": question,
    #"select": "title, product, paragraph",
    #"answers": "extractive",
    #"filter": "category eq 'Databases'",
    #"vectorFilterMode": "postFilter",
    "top": "5"
}

# Convert payload to JSON
data = json.dumps(payload)

# Make the POST request
response_index = requests.post(url, headers=headers, data=data)

# Check if the request was successful
if response_index.status_code == 200:
   
    for item in response_index.json()['value']:
        print(f"Title: {item['title']}")
        print(f"Product: {item['product']}")
        print(f"Paragraph: {item['paragraph']}")
        print(f"Search Score: {item['@search.score']}")
        print()


else:
    print("Request failed with status code:", response_index.status_code)
    print(response_index.text)
```

# Hybrid query with semantic ranking

```python
question = "What are gegenanzeigen for Bevacizumab specified in section 4.1?"

response = client.embeddings.create(
    input = question,
    model= "text-embedding-ada-002"
)

# Define the URL and headers
url = f'https://{search_service_name}.search.windows.net/indexes/{index_name}/docs/search?api-version=2023-11-01'
headers = {
    'Content-Type': 'application/json',
    'api-key': admin_api_key
}

# Define the request body
payload = {
    "vectorQueries": [
        {
            "vector": response.data[0].embedding,
            "fields": "chunkVector",
            "kind": "vector",
            "k": 50
        }
    ],
    "search": question,
    "queryType": "semantic",
    "semanticConfiguration": "my-semantic-config",
    #"select": "title, product, paragraph",
    "captions": "extractive",
    #"answers": "extractive",
    #"filter": "category eq 'Databases'",
    #"vectorFilterMode": "postFilter",
    "top": "5"
}

# Convert payload to JSON
data = json.dumps(payload)

# Make the POST request
response_index = requests.post(url, headers=headers, data=data)

# Check if the request was successful
if response_index.status_code == 200:
   
    for item in response_index.json()['value']:
        print(f"Title: {item['title']}")
        print(f"Product: {item['product']}")
        print(f"Paragraph: {item['paragraph']}")
        print(f"Search Score: {item['@search.score']}")
        print(f"Reranker Score: {item['@search.rerankerScore']}")
        print(f"Reranker Captions: {item['@search.captions']}")
        print()


else:
    print("Request failed with status code:", response_index.status_code)
    print(response_index.text)
```
